## Membuat Graph

```python
edgelist=[
  (1,2),(1,3),(1,4), 
  (2,1), (2,3), 
  (3,1), (3,2), (3,4), 
  (4,5), (4,6), 
  (5,4), (5,6), (5,7), (5,8), 
  (6,5), (6,7), (6,8), 
  (7,5), (7,6), (7,8), (7,9), 
  (8,5), (8,6), (8,7), 
  (9,7)
]
G=nx.Graph(edgelist)
```

```python
nx.draw(G, with_labels = True)
```

## Perhitungan Degree Centrality, Closeness Centrality, Betweenness Centrality

### Degree Centrality

1.  Degree centrality adalah jumlah edge yang terkoneksi pada suatu node yang mewakili interaksi. 
2.  Pentingnya node ditentukan oleh jumlah node yang berdekatan dengan node tersebut 
*   Lebih besar derajatnya (degree), maka lebih penting node itu dalam suatu jaringan 
*   Hanya sebagian kecil node yang memiliki derajat tinggi dalam jaringan

**Degree Centrality**
$$ C_{D}(vi) = d_{i} = \sum_{i}^{}A_{ij} $$
**Normalisasi Degree Centrality**
$$ C'_{D}(vi) = d_{i} = \frac{d_{i}}{(n-1)} $$

```python
dc = nx.degree_centrality(G)
dc
```

### Closeness Centrality

Closenes centrality adalah nilai kedekatan antara satu node dengan node lain dalam jaringan dengan menghitung rata-rata dari jarak relasi node-node tersebut. Skor closeness centrality mewakili kecepatan dalam penyebaran informasi.
$$ C_{u} = \frac{n-1}{\sum_{v=1}^{n-1}d(v,u)} $$

dimana d(v, u) adalah jarak lintasan terpendek antara v dan u, dan n-1 adalah jumlah node yang dapat dijangkau dari u. Perhatikan bahwa fungsi jarak kedekatan menghitung jarak masuk ke u untuk graf-graf berarah.


```python
cc = nx.closeness_centrality(G)
cc
```

### Betweeness Centrality

1.   Skor betweeness Centrality mewakili seberapa besar informasi yang tersebar dari suatu aktor. Semakin besar skor, artinya aktor tersebut semakin berperan dalam penyebaran informasi 
2.   Semakin banyak lintasan yang harus melewati persimpangan itu (misal tidak ada jalan alternatif), maka semakin penting arti persimpangan tersebut. Hal ini menandakan seberapa besar suatu node diperlukan sebagai penghubung dalam penyebaran informasi di dalam jaringan 
3.   Ukuran ini juga dapat digunakan untuk mengidentifikasiÂ **boundary spanners**, yaitu orang atau node yang berperan sebagai penghubung (jembatan) antara dua komunitas

$$ 
\begin{align*}
C_{B} &= \sum_{v_{s} \neq v_{i} \neq v_{t} \in V, s < t}\frac{\sigma_{st}(vi)}{\sigma_{st}} \\
\sigma_{st} &= \text{Jumlah path terpendek antara s dan t} \\
\sigma_{st}(vi) &= \text{Jumlah path terpendek antara s dan t yang melewati }vi 
\end{align*} 
$$

```python
bc = nx.betweenness_centrality(G, normalized=False)
bc
```
