```python
import networkx as nx
```

```python
edgelist=[
  (1,2),(1,3),(1,4), 
  (2,1), (2,3), 
  (3,1), (3,2), (3,4), 
  (4,5), (4,6), 
  (5,4), (5,6), (5,7), (5,8), 
  (6,5), (6,7), (6,8), 
  (7,5), (7,6), (7,8), (7,9), 
  (8,5), (8,6), (8,7), 
  (9,7)
]

G=nx.DiGraph(edgelist)
```

```python
nx.draw_circular(G, with_labels = True)
```

```python
pr = nx.pagerank(G, alpha=0.85, max_iter=25)
pr
```
